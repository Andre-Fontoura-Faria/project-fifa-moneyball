```python
# Removing the players that do not play football anymore:
def contracts(df, column_name):
    df_2021 = df[df[column_name].astype(str).str.contains('2021|2022|2023|2024|2025|2026|2027|2028|2029')]
    return df_2021
```

```python
# The Fifa updated dataset with the relevant columns:
fifa = contracts(fifa, 'Contract')
fifa_2021 = fifa[['Name', 'Age', 'Value','OVA', "Nationality", "Club", 'BP', "Wage", 'Contract']]
```

```python
fifa_2021.head()
```

```python
# Cleaning the Data: 
```

```python
# Cleaning the Value column:

def clean_value(value):
    value = value.replace('€', '')  # remove euro symbol
    if value.endswith('K'):
        value = int(float(value.replace('K', '')) * 1000)  # convert 'K' to 1000 and convert to int
    elif value.endswith('M'):
        value = int(float(value.replace('M', '')) * 1000000)  # convert 'M' to 1000000 and convert to int
    else:
        value = int(value)  # convert to int
    return value

# apply the clean_value function to the 'Value' column
fifa_2021['Value'] = fifa_2021['Value'].apply(clean_value)
```

```python
# Cleaning the Wage column:

def clean_wage(wage):
    wage = wage.replace('€', '')  
    if wage.endswith('K'):
        wage = int(float(wage.replace('K', '')) * 1000)  # convert 'K' to 1000 and convert to int
    elif wage.endswith('M'):
        wage = int(float(wage.replace('M', '')) * 1000000)  # convert 'M' to 1000000 and convert to int
    else:
        wage = int(wage)  # convert to int
    return wage

# apply the clean_wage function to the 'Wage' column
fifa_2021['Wage'] = fifa_2021['Wage'].apply(clean_wage)
```

```python
# Cleaning the Contract column:
```

```python
# Cleaning Contract column - part 1 (including the loans):

fifa_2021 = fifa_2021[~fifa_2021['Contract'].str.contains('On Loan')]
```

```python
# Cleaning Contract column - part 2 (creating 2 new columns):
fifa_2021[['Contract_Begin', 'Contract_End']] = fifa_2021['Contract'].str.split(' ~ ', expand=True)


```

```python
# Cleaning Contract column - part 3 (checking the new columns that still have letters):
non_numeric_rows = fifa_2021[~fifa_2021['Contract_Begin'].str.isdigit() | ~fifa_2021['Contract_End'].str.isdigit()]
print(non_numeric_rows)
```

```python
# Cleaning Contract column - part 4 (replacing the letters from the the column that has letters and convert it to INT):
fifa_2021['Contract_Begin'] = fifa_2021['Contract_Begin'].str.replace(r'\D+', '').astype(int)
```

```python
# Cleaning Contract column - part 5 (convertint the other column to INT):

fifa_2021["Contract_End"] = fifa_2021["Contract_End"].astype(int)
```

```python
# Cleaning Contract column - part 5 (Drop the Contract column):
fifa_final = fifa_2021.drop('Contract', axis=1)

```

```python
# def model to get the best player for each position

def find_best_player(team_data, position):
    position_data = team_data.loc[team_data['BP'] == position]
    max_ova = position_data['OVA'].max()
    best_player = position_data.loc[position_data['OVA'] == max_ova, ['Name', 'OVA']]
    return best_player
```

```python
# Man_City_Future best players for the positions we want for our tactics! 

def find_best_players(team_data, positions):
    best_players = []
    for position in positions:
        position_data = team_data.loc[team_data['BP'] == position]
        max_ova = position_data['OVA'].max()
        best_player = position_data.loc[position_data['OVA'] == max_ova, ['Name', 'OVA']]
        if not best_player.empty:
            best_players.append([position, best_player.iloc[0]['Name'], best_player.iloc[0]['OVA']])
    return pd.DataFrame(best_players, columns=['Position', 'Name', 'OVA'])

criteria = ['GK', 'RWB', 'LWB', 'CB', 'RB', 'CDM', 'CM', 'CAM', 'RW', 'LW', 'ST']

best_players_of_ManCityF = find_best_players(Man_City_Future, criteria)

pd.DataFrame(best_players_of_ManCityF)
```

```python
# Let´s check the Top 10 players in in fifa_final df.

def find_best_players(team_data, positions):
    best_players = []
    for position in positions:
        position_data = team_data.loc[team_data['BP'] == position]
        top_players = position_data.nlargest(10, 'OVA')[['Name', 'OVA', 'Value', 'Age', "Club"]]
        for index, player in top_players.iterrows():
            best_players.append([position, player['Name'], player['OVA'], player['Value'], player['Age'], player['Club']])
    return pd.DataFrame(best_players, columns=['Position', 'Name', 'OVA', 'Value', 'Age', "Club"])

criteria = ['GK', 'RWB', 'LWB', 'CB', 'RB', 'CDM', 'CM', 'CAM', 'RW', 'LW', 'ST']

best_players_df = find_best_players(fifa_final, criteria)

pd.DataFrame(best_players_df)
```
